### We define a function f(value) and we use while loop such that if the function f(x) < 0 and then we print the value of p1.

```python
###define a function to evaluate f(x)
import random
def f(value): 
    return 8*value + 2
    
while (True):
    x = random.randint(-10,10)
    print("trying value ", x)   
    if f(x) < 0 :
        p1 = x
        break
print(p1)
    
```

### IN the same way we use the while loop and generate the values of f(x) such that it is greater than zero and we print the values of p2.

```python

while (True):
    x = random.randint(-10,10)
    print("trying value ", x)   
    if f(x) > 0 :
        p2 = x
        break
print(p2)
```

### we here in this case , we calculate the value p3 as the mid point of p1 and p2 and update either p1 or p2 to this new value p3 after checking if the function is negative or positive and we iterate it till p1=p2 which is the condition we are looking.

```python
p3_values=[]
f_val=[]
while(abs(p2 - p1) > 0.001):
    p3 = (p1 + p2) / 2
    p3_val = f(p3)
    p3_values.append(p3)
    f_val.append(p3_val)
    
    if  p3_val < 0 :
        p1 = p3
        
    else :
        p2 = p3
    
    print(p1, f(p1), p2, f(p2))        

    if(p3_val == 0):
        break
        
print(p1, f(p1), p2, f(p2))   
```
