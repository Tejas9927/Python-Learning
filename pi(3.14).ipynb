calculating the area of sqaure enclosed under a circle by taking the radius as 0.5.

```python
import random
areacircle=0
areasquare=0
x= random.random()/2
y=random.random()/2
print(x,y)
d= (x**2+y**2)**0.5
print(d)
if (d < 0.5):
    areacircle+=1
areasquare+=1    

    
    
```

Finding the value of pi by  using the function append . 

```python
areacircle = 0
areasquare = 0
pi = []
for i in range (0,1000):
    x = random.random()/2
    y = random.random()/2
    d = (x**2 + y**2)**0.5
    if(d < 0.5):
        areacircle += 1 
    areasquare += 1
    pi.append((areacircle/areasquare)*4)    
```

plotting the value of pi by importing matplotlib.

```python
import matplotlib.pyplot as plt
plt.plot(pi)
```

printing the last 10 numbers by using colon.

```python
print(pi[-10:])
```
